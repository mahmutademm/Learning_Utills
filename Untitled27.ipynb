```python
# ================================================================================
# TUTTURO RISK INTELLIGENCE ‚Äî v15.8 (PRECISION PATCH)
# ================================================================================
# CHANGELOG v15.8:
# 1. REMOVED: All "Grade" labels (Elite, Weak, etc.) and text buckets.
# 2. FIXED: 'RatioBenchmarker' now uses np.interp for continuous, precise percentiles.
# 3. TUNING: Curve adjusted so high ratios (7.0) don't get compressed with mid ratios (2.5).
# ================================================================================

import os
import sys
import time
import threading
import numpy as np
import pandas as pd
import yfinance as yf
from flask import Flask, request, render_template_string
from pyngrok import ngrok
from scipy.stats import genpareto, kendalltau

# Check for arch
try:
    from arch import arch_model
except ImportError:
    print("CRITICAL: 'arch' library not found. Please run '!pip install arch'")
    sys.exit()

# ================================================================================
# CONFIGURATION
# ================================================================================

PORT = 5006
app = Flask(__name__)
NGROK_TOKEN = os.getenv("NGROK_TOKEN", "38j35bKIUJape4caraO1Nokp7bt_4jh2zuwp9MtJMyR2fbML5")

SECTOR_BENCHMARKS = {
    "Technology": (45, 1.5, 0.15),
    "Financial Services": (18, 6.0, 0.20),
    "Healthcare": (30, 1.5, 0.10),
    "Energy": (15, 1.2, 0.05),
    "Consumer Cyclical": (25, 2.5, 0.05),
    "Utilities": (20, 4.0, 0.08),
    "Industrials": (22, 2.0, 0.08),
    "Real Estate": (40, 5.0, 0.30),
    "DEFAULT": (25, 2.0, 0.10)
}

# ================================================================================
# THE PRECISE BENCHMARKER (v15.8)
# ================================================================================

class RatioBenchmarker:
    """
    Uses Linear Interpolation to give a precise, continuous percentile.
    No rounding. No buckets. No labels.
    """
    def __init__(self):
        # The 'Anchor Points' of the curve.
        # X = Tutturo Ratio
        # Y = Estimated Market Percentile
        # Tuned to ensure 7.0 (WMT) is clearly separated from 2.5.
        self.ratios =      [0.0, 0.5, 0.8, 1.0, 1.5,  2.0,  3.0,  5.0,  7.0, 10.0]
        self.percentiles = [0.0, 10.0, 30.0, 50.0, 70.0, 80.0, 92.0, 97.0, 99.0, 99.9]

    def get_percentile(self, ratio):
        if ratio < 0: return 0.0
        # Continuous calculation (e.g. 2.5 maps to exactly 86.0%)
        pct = np.interp(ratio, self.ratios, self.percentiles)
        return round(pct, 2) # 2 decimal places precision

    def get_color(self, pct):
        # Purely visual heat map, no text labels
        if pct >= 90: return "#ff00ff" # Magenta (Top 10%)
        if pct >= 75: return "#00f0ff" # Cyan (Top 25%)
        if pct >= 50: return "#00ff00" # Green (Top 50%)
        if pct >= 25: return "#888888" # Grey
        return "#ff2a6d"               # Red (Bottom)

# ================================================================================
# THE NON-LINEAR MACRO GOVERNOR (v15)
# ================================================================================

class MacroGovernor:
    def __init__(self, period="2y"):
        self.tickers = ["^VIX", "^TNX", "CL=F", "DX-Y.NYB"]
        try:
            self.data = yf.download(self.tickers, period=period, progress=False)['Close']
            self.data.index = self.data.index.tz_localize(None)
            self.pct_change = self.data.pct_change().dropna()
        except:
            self.data = pd.DataFrame()

    def _norm(self, val, min_v, max_v):
        score = (val - min_v) / (max_v - min_v) * 100
        return max(1.0, min(100, score))

    def _sigmoid(self, x, L, k, x0):
        return L / (1 + np.exp(-k * (x - x0)))

    def calculate_regime_stress(self):
        if self.data.empty: return {"Score": 50.0, "Regime": 50.0, "Type": "Neutral"}

        try:
            vix_val = self.data["^VIX"] if "^VIX" in self.data else pd.Series([20]*len(self.data))
            tnx_val = self.data["^TNX"] if "^TNX" in self.data else pd.Series([4.0]*len(self.data))
            dxy_val = self.data["DX-Y.NYB"] if "DX-Y.NYB" in self.data else pd.Series([100]*len(self.data))
            oil_val = self.data["CL=F"] if "CL=F" in self.data else pd.Series([70]*len(self.data))
        except:
            return {"Score": 50.0, "Regime": 50.0, "Type": "Neutral"}

        vix_s = vix_val.apply(lambda x: self._norm(x, 12.0, 35.0))
        tnx_s = tnx_val.apply(lambda x: self._norm(x, 3.5, 5.0))
        dxy_s = dxy_val.apply(lambda x: self._norm(x, 95, 110))
        oil_mom = oil_val.pct_change(60).fillna(0)
        oil_s = oil_mom.apply(lambda x: self._norm(x, -0.10, 0.20))

        base = (0.40 * vix_s) + (0.30 * tnx_s) + (0.15 * oil_s) + (0.15 * dxy_s)

        interaction = (vix_s > 50) & (tnx_s > 50)
        final_series = base.copy()
        final_series[interaction] *= 1.25
        final_series = final_series.clip(upper=100)

        regime_series = final_series.ewm(span=14, adjust=False).mean()

        current_score = final_series.iloc[-1]
        regime_score = regime_series.iloc[-1]

        if regime_score > 75: r_type = "CRISIS (Liquidity Trap)"
        elif regime_score > 55: r_type = "FRAGILE (Caution)"
        elif regime_score < 25: r_type = "GOLDILOCKS (Risk On)"
        else: r_type = "NEUTRAL"

        return {"Score": round(current_score, 1), "Regime": round(regime_score, 1), "Type": r_type}

    def get_sigmoid_weights(self, stress_score):
        w_vol = self._sigmoid(stress_score, 0.9, 0.15, 45)
        w_vol = max(0.2, w_vol)

        w_val = 0.3 + (0.3 * (stress_score/100))

        w_mom_penalty = self._sigmoid(stress_score, 1.0, 0.2, 55)
        w_mom = 0.8 * (1.0 - w_mom_penalty)
        w_mom = max(0.1, w_mom)

        w_qual = 0.3 + (0.6 * (stress_score/100))
        w_geo = 0.2 + (0.3 * (stress_score/100))
        w_grow = 0.6 - (0.5 * (stress_score/100))

        return {
            "R_VOL": w_vol, "R_VAL": w_val, "R_GEO": w_geo,
            "O_MOM": w_mom, "O_GROW": w_grow, "O_QUAL": w_qual
        }

    def get_stock_modifier(self, stock_rets, stress_score):
        if self.data.empty or "^VIX" not in self.pct_change: return 1.0, "Neutral"

        df = stock_rets.to_frame('S').join(self.pct_change['^VIX'], how='inner')
        if df.empty: return 1.0, "Neutral"
        corr_vix = df.corr().iloc[0,1]

        hedge_efficacy = 1.0
        if stress_score > 60:
            hedge_efficacy = max(0, 1.0 - ((stress_score - 60) / 40.0))

        base_mod = 1.0
        msg = "Neutral"

        if corr_vix > 0.2:
            penalty = 1.0 + (corr_vix * (stress_score/100.0))
            base_mod = penalty
            msg = f"Systemic Risk (Corr {corr_vix:.2f})"
        elif corr_vix < -0.2:
            raw_discount = abs(corr_vix) * 0.5 * hedge_efficacy
            discount = min(0.25, raw_discount)
            base_mod = 1.0 - discount

            if hedge_efficacy < 0.2:
                msg = f"HEDGE FAILED (Liquidity Crunch)"
                base_mod = 1.1
            else:
                msg = f"Hedge (Eff: {hedge_efficacy:.0%})"

        return round(base_mod, 2), msg

# ================================================================================
# THE MASTER ENGINE (v15.6)
# ================================================================================

class TutturoMasterEngine:
    def __init__(self, ticker, period="5y"):
        self.ticker = ticker.upper()
        self.stock = yf.Ticker(self.ticker)
        self.governor = MacroGovernor(period="2y")
        self.benchmarker = RatioBenchmarker()

        try:
            self.info = self.stock.info
            self.sector = self.info.get("sector", "DEFAULT") if self.info else "DEFAULT"
        except: self.info = {}; self.sector = "DEFAULT"

        data = yf.download([self.ticker, "SPY"], period=period, group_by='ticker', progress=False)
        try:
            self.history = data[self.ticker]
            self.spy_history = data["SPY"]
        except KeyError:
            try:
                self.history = yf.download(self.ticker, period=period, progress=False)
                self.spy_history = yf.download("SPY", period=period, progress=False)
            except:
                raise ValueError("No Price Data Found")

        if self.history.empty: raise ValueError("No Price Data Found")

        self.history.index = self.history.index.tz_localize(None)
        self.spy_history.index = self.spy_history.index.tz_localize(None)

        self.returns = self.history["Close"].pct_change().dropna()
        self.spy_returns = self.spy_history["Close"].pct_change().dropna()
        self.aligned_data = pd.DataFrame({'Asset': self.returns, 'Market': self.spy_returns}).dropna()

        self.b_pe, self.b_de, self.b_margin = SECTOR_BENCHMARKS.get(self.sector, SECTOR_BENCHMARKS["DEFAULT"])

    def _scale_risk(self, val, safe, danger):
        if danger == safe: return 50.0
        score = (val - safe) / (danger - safe) * 100.0
        return max(1.0, min(100.0, score))

    def _scale_opp(self, val, bad, good):
        if good == bad: return 0.0
        score = (val - bad) / (good - bad) * 100.0
        return max(0.0, min(100.0, score))

    def get_raw_volatility(self):
        try:
            r_scaled = self.returns * 100
            am = arch_model(r_scaled, vol='Garch', p=1, q=1, dist='Normal')
            res = am.fit(disp='off', show_warning=False)
            garch_vol = (np.sqrt(res.forecast(horizon=1).variance.iloc[-1, 0])/100) * np.sqrt(252)
        except: garch_vol = self.returns.std() * np.sqrt(252)

        losses = -1 * self.returns
        threshold = np.percentile(losses, 95)
        excesses = losses[losses > threshold] - threshold
        try:
            shape, loc, scale = genpareto.fit(excesses)
            nu, n = len(excesses), len(losses)
            evt_var = threshold + (scale / shape) * ( ((n / nu) * (1 - 0.99))**(-shape) - 1 )
        except: evt_var = np.percentile(losses, 99)

        try:
            tau, _ = kendalltau(self.aligned_data['Asset'], self.aligned_data['Market'])
            theta = (2*tau)/(1-tau) if tau > 0 else 0
            tail_dep = 2**(-1/theta) if theta > 0 else 0
        except: tail_dep = 0

        self.atoms_vol = {"GARCH": garch_vol, "EVT": evt_var, "TAIL": tail_dep}

        return {
            "GARCH": self._scale_risk(garch_vol, 0.15, 0.60),
            "EVT": self._scale_risk(evt_var, 0.05, 0.15),
            "TAIL": self._scale_risk(tail_dep, 0.1, 0.6),
            "VAL_VAR": evt_var
        }

    def get_raw_fundamentals(self):
        de = self.info.get("debtToEquity", self.b_de*100) or self.b_de*100; de = de/100.0
        pe = self.info.get("trailingPE", 25) or 25
        rev_g = self.info.get("revenueGrowth", 0) or 0
        earn_g = self.info.get("earningsGrowth", 0) or 0

        if earn_g and earn_g > 0:
            peg = pe / (earn_g * 100)
        else:
            peg = 5.0

        self.atoms_fund = {"DE": de, "PE": pe, "REV_G": rev_g, "EARN_G": earn_g, "PEG": peg}

        return {
            "STABILITY": self._scale_risk(de, self.b_de*0.5, self.b_de*1.5),
            "VALUATION_R": self._scale_risk(pe, self.b_pe*0.8, self.b_pe*2.0),
            "REV_G": self._scale_opp(rev_g, 0.0, 0.30),
            "EARN_G": self._scale_opp(earn_g, 0.0, 0.30),
            "PEG": 100 - self._scale_risk(peg, 0.5, 3.0),
            "VALUATION_O": 100 - self._scale_risk(pe, self.b_pe*0.5, self.b_pe*1.5)
        }

    def get_raw_technical(self):
        curr = self.history["Close"].iloc[-1]
        ma50 = self.history["Close"].rolling(50).mean().iloc[-1]
        mom_pct = (curr - ma50) / ma50

        target = self.info.get("targetMeanPrice", curr)
        if target is None: target = curr
        up_pct = (target - curr) / curr

        stock_ret = (curr / self.history["Close"].iloc[0]) - 1
        spy_ret = (self.spy_history["Close"].iloc[-1] / self.spy_history["Close"].iloc[0]) - 1
        alpha_val = stock_ret - spy_ret

        self.atoms_tech = {"MOM": mom_pct, "UP": up_pct, "ALPHA": alpha_val}

        return {
            "MOMENTUM": self._scale_opp(mom_pct, -0.10, 0.20),
            "ANALYST": self._scale_opp(up_pct, 0.0, 0.50),
            "ALPHA": self._scale_opp(alpha_val, -0.20, 0.20)
        }

    def _audit(self, name, raw, score, weight, source):
        if abs(raw) < 5.0 and name not in ["PEG Ratio", "Valuation (P/E)"]:
            raw_fmt = f"{raw:.2%}"
        else:
            raw_fmt = f"{raw:.2f}"
        return {"Name": name, "Raw": raw_fmt, "Score": round(score,1), "Weight": round(weight,2), "Net": round(score*weight, 1), "Source": source}

    def run_horizon_analysis(self):
        vol = self.get_raw_volatility()
        fund = self.get_raw_fundamentals()
        tech = self.get_raw_technical()

        macro = self.governor.calculate_regime_stress()
        stress = macro["Regime"]
        w = self.governor.get_sigmoid_weights(stress)
        mod_val, mod_msg = self.governor.get_stock_modifier(self.returns, stress)

        # === SHORT TERM PIPELINE ===
        audit_rs = [
            self._audit("GARCH Volatility", self.atoms_vol["GARCH"], vol['GARCH'], w["R_VOL"], "ARCH Model (GARCH(1,1))"),
            self._audit("EVT Tail Risk", self.atoms_vol["EVT"], vol['EVT'], (1-w["R_VOL"]), "Gen Pareto Distribution (95% Tail)")
        ]
        r_short_raw = sum([x["Net"] for x in audit_rs])
        r_short = max(5.0, r_short_raw * mod_val)

        audit_os = [
            self._audit("Momentum (vs 50MA)", self.atoms_tech["MOM"], tech['MOMENTUM'], w["O_MOM"], "Price vs 50-Day MA"),
            self._audit("Analyst Upside", self.atoms_tech["UP"], tech['ANALYST'], (1-w["O_MOM"]), "Target Price vs Current")
        ]
        o_short = sum([x["Net"] for x in audit_os])

        # === MEDIUM TERM PIPELINE ===
        audit_rm = [
            self._audit("Valuation (P/E)", self.atoms_fund["PE"], fund['VALUATION_R'], w["R_VAL"], "Trailing P/E Ratio"),
            self._audit("GARCH Volatility", self.atoms_vol["GARCH"], vol['GARCH'], w["R_VOL"], "ARCH Model (GARCH(1,1))"),
            self._audit("Fin Stability (D/E)", self.atoms_fund["DE"], fund['STABILITY'], 0.2, "Debt / Equity Ratio")
        ]
        r_med_raw = sum([x["Net"] for x in audit_rm])
        r_med = max(5.0, r_med_raw)

        audit_om = [
            self._audit("Earnings Growth", self.atoms_fund["EARN_G"], fund['EARN_G'], w["O_GROW"], "YoY Earnings Growth"),
            self._audit("PEG Ratio", self.atoms_fund["PEG"], fund['PEG'], 0.3, "P/E divided by Growth"),
            self._audit("Momentum", self.atoms_tech["MOM"], tech['MOMENTUM'], 0.2, "Price vs 50-Day MA")
        ]
        o_med = sum([x["Net"] for x in audit_om])

        # === LONG TERM PIPELINE ===
        audit_rl = [
            self._audit("Fin Stability (D/E)", self.atoms_fund["DE"], fund['STABILITY'], 0.5, "Debt / Equity Ratio"),
            self._audit("Tail Dependency", self.atoms_vol["TAIL"], vol['TAIL'], w["R_GEO"], "Clayton Copula (Kendall's Tau)"),
            self._audit("Valuation (P/E)", self.atoms_fund["PE"], fund['VALUATION_R'], 0.2, "Trailing P/E Ratio")
        ]
        r_long_raw = sum([x["Net"] for x in audit_rl])
        r_long = max(5.0, r_long_raw)

        audit_ol = [
            self._audit("Alpha (vs SPY)", self.atoms_tech["ALPHA"], tech['ALPHA'], w["O_QUAL"], "Cumulative Return Alpha"),
            self._audit("Revenue Growth", self.atoms_fund["REV_G"], fund['REV_G'], 0.3, "YoY Revenue Growth"),
            self._audit("PEG Ratio", self.atoms_fund["PEG"], fund['PEG'], 0.2, "P/E divided by Growth")
        ]
        o_long = sum([x["Net"] for x in audit_ol])

        unprofit = self.check_profitability()
        if unprofit:
            r_short += 15; r_med += 25; r_long += 35
            profit_msg = "YES (+Penalty)"
        else: profit_msg = "NO"

        # === CALCULATE PERCENTILES (NO LABELS) ===
        ratio_s = o_short / r_short
        ratio_m = o_med / r_med
        ratio_l = o_long / r_long

        pct_s = self.benchmarker.get_percentile(ratio_s)
        pct_m = self.benchmarker.get_percentile(ratio_m)
        pct_l = self.benchmarker.get_percentile(ratio_l)

        results = {
            "Short": {
                "R": round(min(99, r_short),1), "O": round(o_short,1),
                "Ratio": round(ratio_s, 2), "Pct": pct_s, "Color": self.benchmarker.get_color(pct_s)
            },
            "Medium": {
                "R": round(min(99, r_med),1), "O": round(o_med,1),
                "Ratio": round(ratio_m, 2), "Pct": pct_m, "Color": self.benchmarker.get_color(pct_m)
            },
            "Long": {
                "R": round(min(99, r_long),1), "O": round(o_long,1),
                "Ratio": round(ratio_l, 2), "Pct": pct_l, "Color": self.benchmarker.get_color(pct_l)
            },
            "Gov": {"Score": macro["Score"], "RegimeScore": macro["Regime"], "Type": macro["Type"], "Msg": mod_msg, "Mod": mod_val, "Weights": w},
            "Audit": {
                "Short_R": audit_rs, "Short_O": audit_os, "Short_Calc": {"Raw": round(r_short_raw,1), "Mod": mod_val},
                "Med_R": audit_rm, "Med_O": audit_om,
                "Long_R": audit_rl, "Long_O": audit_ol,
                "Profit": profit_msg
            },
            "EVT_VaR": vol['VAL_VAR']
        }
        return results

    def check_profitability(self):
        eps = self.info.get("trailingEps")
        return True if (eps is None or eps < 0) else False

# ================================================================================
# FLASK UI
# ================================================================================

HTML = r"""
<!DOCTYPE html>
<html>
<head>
<title>Tutturo v15.8</title>
<script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
<link href="https://fonts.googleapis.com/css2?family=JetBrains+Mono:wght@400;800&family=Inter:wght@400;700;900&display=swap" rel="stylesheet">
<script type="text/x-mathjax-config">
MathJax.Hub.Config({
  tex2jax: {inlineMath: [['\\(','\\)']], processEscapes: true}
});
</script>
<script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-AMS_HTML" async></script>
<style>
    :root { --bg: #050505; --card: #111; --green: #00f0ff; --red: #ff2a6d; --text: #eee; --gold: #ffd700; }
    body { background: var(--bg); color: var(--text); font-family: 'Inter', sans-serif; padding: 40px; }
    .container { max-width: 1300px; margin: 0 auto; }
    .brand { font-size: 30px; font-weight: 900; margin-bottom: 5px; }
    .brand span { color: var(--green); }

    input { background: #000; border: 1px solid #333; color: var(--green); padding: 15px; font-family: 'JetBrains Mono'; width: 200px; font-weight:bold; }
    button { background: var(--green); border: none; padding: 15px 30px; font-weight: 900; cursor: pointer; }
    select { background: #222; color: #fff; border: 1px solid #444; padding: 15px; font-weight:bold; margin-left:10px; cursor:pointer;}

    .grid-main { display: grid; grid-template-columns: 1fr 1fr; gap: 30px; margin-top: 30px; }
    .card { background: var(--card); border: 1px solid #222; padding: 25px; border-radius: 6px; }
    .score-title { font-family: 'JetBrains Mono'; font-size: 13px; color: #888; letter-spacing: 1px; text-transform: uppercase; margin-bottom: 10px; }
    .big-num { font-size: 80px; font-weight: 900; line-height: 1; }

    .pct-display {
        display: inline-block;
        font-family: 'JetBrains Mono';
        font-size: 16px;
        font-weight: bold;
        margin-bottom: 15px;
    }

    .risk-high { color: var(--red); } .risk-low { color: var(--green); }
    .opp-high { color: var(--gold); } .opp-low { color: #555; }
    .view-section { display: none; }
    .view-active { display: block; }

    /* GAUGE BAR */
    .gauge-container { width: 100%; height: 10px; background: #333; border-radius: 5px; margin-top: 10px; position: relative; }
    .gauge-fill { height: 100%; background: linear-gradient(90deg, #00f0ff, #ffd700, #ff2a6d); border-radius: 5px; transition: width 1s ease; }
    .gauge-marker { position: absolute; top: -5px; width: 4px; height: 20px; background: #fff; border: 1px solid #000; }

    /* PIPELINE STYLES */
    details { margin-top: 40px; border: 1px solid #333; border-radius: 6px; padding: 10px; background: #0a0a0a; }
    summary { font-family: 'JetBrains Mono'; cursor: pointer; color: var(--gold); font-weight: bold; padding: 10px; outline: none; }
    .step-box { border-left: 2px solid #333; margin-left: 10px; padding-left: 20px; padding-bottom: 20px; }
    .step-title { font-family:'JetBrains Mono'; color: #fff; font-weight:bold; margin-bottom:10px; }
    .step-desc { font-size:12px; color:#888; margin-bottom:10px; }
    table { width: 100%; border-collapse: collapse; margin-top: 10px; font-family: 'JetBrains Mono'; font-size: 11px; }
    th { text-align: left; border-bottom: 1px solid #444; padding: 5px; color: #888; }
    td { border-bottom: 1px solid #222; padding: 5px; }
</style>
<script>
    function switchView() {
        var selector = document.getElementById("horizonSelect");
        var val = selector.value;
        document.getElementById("Short").classList.remove("view-active");
        document.getElementById("Medium").classList.remove("view-active");
        document.getElementById("Long").classList.remove("view-active");
        document.getElementById(val).classList.add("view-active");

        document.getElementById("ExpShort").style.display = (val === "Short") ? "block" : "none";
        document.getElementById("ExpMedium").style.display = (val === "Medium") ? "block" : "none";
        document.getElementById("ExpLong").style.display = (val === "Long") ? "block" : "none";
    }
</script>
</head>
<body>
<div class="container">
    <div style="border-bottom:2px solid #222; padding-bottom:20px; margin-bottom:30px;">
        <div class="brand">TUTTURO<span>v15.8</span></div>
        <small style="color:#666; font-family:'JetBrains Mono'">CAUSAL PIPELINE ENGINE (PRECISION PATCH)</small>
    </div>

    <form method="POST">
        <input name="ticker" placeholder="TICKER" required value="{{ t }}">
        <button type="submit">ANALYZE</button>
    </form>

    {% if res %}
    <div style="margin-top: 30px; background: #1a1a1a; padding: 20px; border-radius: 8px;">
        <div style="display:flex; justify-content:space-between; align-items:flex-end;">
            <div class="score-title">MACRO REGIME: <span style="color:#fff">{{ data['Gov']['Type'] }}</span></div>
            <div style="font-family:'JetBrains Mono'; font-weight:bold; color: #fff;">{{ data['Gov']['RegimeScore'] }}/100</div>
        </div>
        <div class="gauge-container">
            <div class="gauge-fill" style="width: 100%;"></div>
            <div class="gauge-marker" style="left: {{ data['Gov']['RegimeScore'] }}%;"></div>
        </div>
        <div style="display:flex; justify-content:space-between; margin-top:5px; font-size:11px; color:#666; font-family:'JetBrains Mono'">
            <span>GOLDILOCKS (0)</span><span>NEUTRAL (50)</span><span>LIQUIDITY CRISIS (100)</span>
        </div>
    </div>

    <div style="margin-top: 20px; border-bottom: 1px solid #333; padding-bottom: 20px;">
        <label style="font-family:'JetBrains Mono'; color:#888;">TIME HORIZON:</label>
        <select id="horizonSelect" onchange="switchView()">
            <option value="Short">SHORT (Tactical | 1-3 Mo)</option>
            <option value="Medium">MEDIUM (Strategic | 1 Yr)</option>
            <option value="Long">LONG (Fundamental | 5 Yr)</option>
        </select>
        <div style="margin-top:15px; font-family:'JetBrains Mono'; font-size:14px; color:#888;">
            RATIOS:
            <span style="color:var(--green); margin-right:15px;">Short: {{ data['Short']['Ratio'] }}x</span>
            <span style="color:var(--gold); margin-right:15px;">Med: {{ data['Medium']['Ratio'] }}x</span>
            <span style="color:#fff;">Long: {{ data['Long']['Ratio'] }}x</span>
        </div>
    </div>

    <div id="Short" class="view-section view-active">
        <div class="pct-display" style="color: {{ data['Short']['Color'] }}">PERCENTILE: {{ data['Short']['Pct'] }}%</div>
        <div class="grid-main">
            <div class="card" style="border-top: 4px solid var(--red);">
                <div class="score-title">TACTICAL RISK</div>
                <div class="big-num {{ 'risk-high' if data['Short']['R'] > 50 else 'risk-low' }}">{{ data['Short']['R'] }}</div>
            </div>
            <div class="card" style="border-top: 4px solid var(--gold);">
                <div class="score-title">TACTICAL OPPORTUNITY</div>
                <div class="big-num {{ 'opp-high' if data['Short']['O'] > 50 else 'opp-low' }}">{{ data['Short']['O'] }}</div>
            </div>
        </div>
    </div>

    <div id="Medium" class="view-section">
        <div class="pct-display" style="color: {{ data['Medium']['Color'] }}">PERCENTILE: {{ data['Medium']['Pct'] }}%</div>
        <div class="grid-main">
            <div class="card" style="border-top: 4px solid var(--red);">
                <div class="score-title">STRATEGIC RISK</div>
                <div class="big-num {{ 'risk-high' if data['Medium']['R'] > 50 else 'risk-low' }}">{{ data['Medium']['R'] }}</div>
            </div>
            <div class="card" style="border-top: 4px solid var(--gold);">
                <div class="score-title">STRATEGIC OPP</div>
                <div class="big-num {{ 'opp-high' if data['Medium']['O'] > 50 else 'opp-low' }}">{{ data['Medium']['O'] }}</div>
            </div>
        </div>
    </div>

    <div id="Long" class="view-section">
        <div class="pct-display" style="color: {{ data['Long']['Color'] }}">PERCENTILE: {{ data['Long']['Pct'] }}%</div>
        <div class="grid-main">
            <div class="card" style="border-top: 4px solid var(--red);">
                <div class="score-title">FUNDAMENTAL RISK</div>
                <div class="big-num {{ 'risk-high' if data['Long']['R'] > 50 else 'risk-low' }}">{{ data['Long']['R'] }}</div>
            </div>
            <div class="card" style="border-top: 4px solid var(--gold);">
                <div class="score-title">FUNDAMENTAL OPP</div>
                <div class="big-num {{ 'opp-high' if data['Long']['O'] > 50 else 'opp-low' }}">{{ data['Long']['O'] }}</div>
            </div>
        </div>
    </div>

    <details>
        <summary>‚ñº CAUSAL CALCULATION GRAPH (EXPLAINABLE AI)</summary>
        <div style="padding: 20px;">

            <div class="step-box" style="border-left-color: #555;">
                <div class="step-title">PHASE 1: BASE DATA ATOMS</div>
                <div class="step-desc">Raw financial data extracted from market feeds.</div>
                <div id="ExpShort">
                    <table>
                        <thead><tr><th>METRIC</th><th>RAW VALUE</th><th>SOURCE</th></tr></thead>
                        <tbody>
                            {% for row in data['Audit']['Short_R'] %}
                            <tr><td>{{ row['Name'] }}</td><td style="color:var(--green)">{{ row['Raw'] }}</td><td>{{ row['Source'] }}</td></tr>
                            {% endfor %}
                            {% for row in data['Audit']['Short_O'] %}
                            <tr><td>{{ row['Name'] }}</td><td style="color:var(--green)">{{ row['Raw'] }}</td><td>{{ row['Source'] }}</td></tr>
                            {% endfor %}
                        </tbody>
                    </table>
                </div>
                <div id="ExpMedium" style="display:none">
                    <table>
                        <thead><tr><th>METRIC</th><th>RAW VALUE</th><th>SOURCE</th></tr></thead>
                        <tbody>
                            {% for row in data['Audit']['Med_R'] %}
                            <tr><td>{{ row['Name'] }}</td><td style="color:var(--green)">{{ row['Raw'] }}</td><td>{{ row['Source'] }}</td></tr>
                            {% endfor %}
                            {% for row in data['Audit']['Med_O'] %}
                            <tr><td>{{ row['Name'] }}</td><td style="color:var(--green)">{{ row['Raw'] }}</td><td>{{ row['Source'] }}</td></tr>
                            {% endfor %}
                        </tbody>
                    </table>
                </div>
                <div id="ExpLong" style="display:none">
                    <table>
                        <thead><tr><th>METRIC</th><th>RAW VALUE</th><th>SOURCE</th></tr></thead>
                        <tbody>
                            {% for row in data['Audit']['Long_R'] %}
                            <tr><td>{{ row['Name'] }}</td><td style="color:var(--green)">{{ row['Raw'] }}</td><td>{{ row['Source'] }}</td></tr>
                            {% endfor %}
                            {% for row in data['Audit']['Long_O'] %}
                            <tr><td>{{ row['Name'] }}</td><td style="color:var(--green)">{{ row['Raw'] }}</td><td>{{ row['Source'] }}</td></tr>
                            {% endfor %}
                        </tbody>
                    </table>
                </div>
            </div>

            <div class="step-box" style="border-left-color: var(--gold);">
                <div class="step-title">PHASE 2: MACRO GOVERNOR PHYSICS</div>
                <div class="step-desc">Macro Stress (S) warps the calculation weights using Sigmoid functions.</div>
                <div style="background:#111; padding:10px; font-family:'JetBrains Mono'; font-size:11px; color:#aaa;">
                    S = {{ data['Gov']['RegimeScore'] }} ({{ data['Gov']['Type'] }})<br><br>
                    w_vol = œÉ(S; L=0.9, k=0.15, x0=45) = <b style="color:#fff">{{ "%.2f"|format(data['Gov']['Weights']['R_VOL']) }}</b><br>
                    w_val = 0.3 + (0.3 * (S/100)) = <b style="color:#fff">{{ "%.2f"|format(data['Gov']['Weights']['R_VAL']) }}</b><br>
                    w_geo = 0.2 + (0.3 * (S/100)) = <b style="color:#fff">{{ "%.2f"|format(data['Gov']['Weights']['R_GEO']) }}</b><br>
                    w_mom = 0.8 * (1 - œÉ(S; L=1.0, k=0.2, x0=55)) = <b style="color:#fff">{{ "%.2f"|format(data['Gov']['Weights']['O_MOM']) }}</b><br>
                    w_grow = 0.6 - (0.5 * (S/100)) = <b style="color:#fff">{{ "%.2f"|format(data['Gov']['Weights']['O_GROW']) }}</b><br>
                    w_qual = 0.3 + (0.6 * (S/100)) = <b style="color:#fff">{{ "%.2f"|format(data['Gov']['Weights']['O_QUAL']) }}</b>
                </div>
            </div>

            <div class="step-box" style="border-left-color: var(--green);">
                <div class="step-title">PHASE 3: WEIGHTED TRANSFORMATION (PATCHED)</div>
                <div class="step-desc">Explicit Vector Scaling: _scale_risk(0-100) and _scale_opp(0-100).</div>
                <div style="background:#111; padding:10px; font-family:'JetBrains Mono'; font-size:11px; color:#aaa; margin-top:10px;">
                    Risk Metrics: Safe Limit ‚Üí Danger Limit (0 ‚Üí 100)<br>
                    Opp Metrics: Bad Limit ‚Üí Good Limit (0 ‚Üí 100)<br>
                </div>
                <table>
                    <thead><tr><th>METRIC</th><th>NORM SCORE</th><th>WEIGHT</th><th>NET IMPACT</th></tr></thead>
                    <tbody>
                        <tr><td colspan="4" style="text-align:center; color:#555;">(See specific horizon tables above for details)</td></tr>
                    </tbody>
                </table>
            </div>

            <div class="step-box" style="border-left-color: var(--red);">
                <div class="step-title">PHASE 4: FINAL ASSEMBLY</div>
                <div class="step-desc">Aggregating impacts and applying the Physics Floor.</div>
                <div style="background:#111; padding:10px; font-family:'JetBrains Mono'; font-size:11px;">
                    RISK_RAW = SUM(Impacts)<br>
                    MODIFIER = {{ data['Gov']['Mod'] }}x ({{ data['Gov']['Msg'] }})<br>
                    {% if data['Audit']['Profit'] == "YES (+Penalty)" %}
                    PENALTY = +15 (Short) / +25 (Med) / +35 (Long) [Unprofitable]<br>
                    {% endif %}
                    FLOOR = Max(5.0, RISK_RAW * MODIFIER)<br><br>
                    <span style="color:var(--green)">FINAL RATIO = OPPORTUNITY / FLOOR</span>
                </div>
            </div>

        </div>
    </details>

    <div class="card" style="margin-top:20px; border-color:var(--red);">
        <div class="score-title" style="color:var(--red)">INSTITUTIONAL CRASH LINE (EVT 99%)</div>
        <div style="font-size:20px; font-weight:bold;">-${{ "%.2f"|format(evt_dollar) }} per share</div>
    </div>

    <div style="margin-top:50px; border-top:1px solid #333; padding-top:20px; font-size:10px; color:#444; font-family:'Inter'">
        DISCLAIMER: Tutturo is an educational research tool. The Risk/Opportunity scores generated are derived from mathematical formulas and do not constitute financial advice, investment recommendations, or an offer to sell securities. Past performance of the algorithm does not guarantee future results.
    </div>

    {% endif %}
</div>
</body>
</html>
"""

@app.route("/", methods=["GET", "POST"])
def index():
    if request.method == "POST":
        try:
            t = request.form["ticker"]
            eng = TutturoMasterEngine(t)
            data = eng.run_horizon_analysis()
            curr_price = eng.history["Close"].iloc[-1]
            evt_dollar = curr_price * data['EVT_VaR']
            return render_template_string(HTML, res=True, t=t, data=data, evt_dollar=evt_dollar)
        except Exception as e: return f"<h1 style='color:red; font-family:sans-serif'>Error: {e}</h1>"
    return render_template_string(HTML, res=False, t="")

def run():
    app.run(port=PORT, use_reloader=False)

if __name__ == "__main__":
    os.system("pkill -9 -f ngrok")
    threading.Thread(target=run, daemon=True).start()
    time.sleep(1)
    if NGROK_TOKEN:
        ngrok.set_auth_token(NGROK_TOKEN)
        print(f"\nüöÄ TUTTURO v15.8 LIVE: {ngrok.connect(PORT).public_url}\n")
    else: print("Running Local")
    while True: time.sleep(1)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from tqdm import tqdm
from arch import arch_model
from scipy.stats import genpareto
import warnings
warnings.filterwarnings("ignore")

# =============================
# CONFIG
# =============================
START = "2014-01-01"
END   = "2024-01-01"
TOP_N = 20
REB_FREQ = "M"

# =============================
# HARD UNIVERSE (NO WIKI)
# =============================
tickers = [
"AAPL","MSFT","AMZN","NVDA","GOOGL","META","BRK-B","LLY","AVGO","TSLA","JPM","V",
"XOM","UNH","MA","COST","PG","HD","JNJ","ORCL","MRK","CRM","ABBV","CVX","KO","PEP",
"ADBE","BAC","NFLX","AMD","INTC","TMO","ACN","WMT","MCD","DIS","CSCO","ABT","LIN",
"DHR","PFE","TXN","NKE","QCOM","PM","UPS","HON","LOW","BMY","SBUX","MDT","INTU",
"RTX","IBM","AMGN","ELV","CAT","GE","GS","DE","BLK","SYK","NOW","LMT","SPGI","ISRG",
"ADI","CB","PLD","ZTS","CI","TGT","BDX","PNC","USB","C","BK","APD","GILD","MO"
]

# =============================
# BENCHMARK
# =============================
spy = yf.download("SPY", start=START, end=END, auto_adjust=True)["Close"]

# =============================
# REBALANCE DATES
# =============================
dates = pd.date_range(START, END, freq=REB_FREQ)

portfolio_curve = []
benchmark_curve = []

portfolio_value = 1.0
benchmark_value = 1.0

# =============================
# CORE SIGNAL COMPONENTS
# =============================
def garch_vol(returns):
    am = arch_model(returns*100, vol='Garch', p=1, q=1)
    res = am.fit(disp="off")
    return np.sqrt(252) * res.conditional_volatility.iloc[-1] / 100

def evt_tail(returns):
    losses = -returns
    u = np.quantile(losses, 0.95)
    excess = losses[losses > u] - u
    if len(excess) < 10:
        return np.nan
    xi, beta = genpareto.fit(excess)
    return u + (beta/xi)*(((len(losses)/(len(excess)))*(1-0.99))**(-xi)-1)

def momentum(prices):
    ma = prices.rolling(50).mean().iloc[-1]
    return (prices.iloc[-1] - ma) / ma

# =============================
# TUTTURO SCORE
# =============================
def tutturo_score(ticker):
    px = yf.download(ticker, period="3y", auto_adjust=True, progress=False)["Close"]
    if len(px) < 200:
        return None

    r = px.pct_change().dropna()

    vol = garch_vol(r)
    tail = evt_tail(r)
    mom = momentum(px)

    upside = np.random.uniform(0.05,0.3)  # proxy

    # normalize (simple z-score proxy)
    risk = (vol*50 + tail*10)
    opp  = (mom*100 + upside*100)

    return opp / max(5, risk)

# =============================
# MAIN LOOP
# =============================
for i in tqdm(range(len(dates)-1)):
    t0 = dates[i]
    t1 = dates[i+1]

    scores = []

    for tk in tickers:
        try:
            s = tutturo_score(tk)
            if s:
                scores.append((tk,s))
        except:
            pass

    if len(scores) < TOP_N:
        continue

    top = pd.DataFrame(scores, columns=["Ticker","Score"]) \
            .sort_values("Score", ascending=False).head(TOP_N)

    w = 1.0 / TOP_N
    period_ret = 0

    for tk in top["Ticker"]:
        try:
            px = yf.download(tk, start=t0, end=t1, auto_adjust=True, progress=False)["Close"]
            r = (px.iloc[-1] / px.iloc[0]) - 1
            period_ret += w * r
        except:
            pass

    portfolio_value *= (1 + period_ret)
    portfolio_curve.append(portfolio_value)

    spy_px = spy.loc[t0:t1]
    if len(spy_px) > 1:
        benchmark_value *= (spy_px.iloc[-1] / spy_px.iloc[0])

    benchmark_curve.append(benchmark_value)

# =============================
# RESULTS
# =============================
bt = pd.DataFrame({
    "Date": dates[1:len(portfolio_curve)+1],
    "TUTTURO": portfolio_curve,
    "SPY": benchmark_curve
}).set_index("Date")

bt.plot(title="TUTTURO Top-20 vs SPY", figsize=(12,6))
```
